# Data Validation

The dataset contains **15000 rows** and **8 columns** before data validation and cleaning. After thorough validation and cleaning, we fixed the dataset to **15000 rows** and **8 columns** ensuring data integrity and reliability for our analysis.

1. Imputed 1074 missing rows in revenue column with the help of median for each sales category. Since the missing data made up more than 5% of data, I imputed it. Median was selected as imputation method over mean, to avoid getting fualty data due to outliers.
2.  Data was cleaned by fixing 2 wrong categories('email' , 'em + call'). 3 categories('EMAIL', 'CALL','EMAIL + CALL') were obtained in the end.
3.  Imputed outliers in the 'years_as_customer' column (values > 39 years) with the max value of 39 years.

# Business Metric

Our goal is to optimize sales strategies for the newly launced office products, while considering revenue and efficiency in mind. I would recommend to use **Revenue generated per week in the last 6 weeks as our key metric**

This metric allows us 
1. To track performance of each method over the duration.
2. To understand trends in sales performance with respect to the effort and cost spent.
3. Make data driven decisions  on resource allocation.

In the last 6 weeks the revenue generation has increased for all categories, from around 40% for Calls only approach, which is our poorest performing method, to around 62% for Calls and Emails combined approach, which is performing best.
